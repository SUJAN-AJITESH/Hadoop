```python
import sys

def mapper():
    for line in sys.stdin:
        parts = line.strip().split("\t")
        node = parts[0]
        adjacency_list = parts[1] if parts[1] != "null" else "[]"
        distance = int(parts[2]) if parts[2] != "∞" else float("inf")
        backpointer = parts[3] if parts[3] != "null" else None

        print(f"{node}\t{adjacency_list}\t{distance}\t{backpointer}")

        if distance != float("inf"):
            neighbors = eval(adjacency_list)
            for neighbor in neighbors:
                print(f"{neighbor}\tnull\t{distance + 1}\t{node}")

if __name__ == "__main__":
    mapper()
```

```python
import sys

def reducer():
    graph = {}

    for line in sys.stdin:
        parts = line.strip().split("\t")
        node = parts[0]
        adjacency_list = parts[1] if parts[1] != "null" else None
        distance = int(parts[2]) if parts[2] != "∞" else float("inf")
        backpointer = parts[3] if parts[3] != "null" else None

        if node not in graph:
            graph[node] = {"adjacency_list": "[]", "distance": float("inf"), "backpointer": None}

        if adjacency_list:
            graph[node]["adjacency_list"] = adjacency_list

        if distance < graph[node]["distance"]:
            graph[node]["distance"] = distance
            graph[node]["backpointer"] = backpointer

    for node, data in graph.items():
        print(f"{node}\t{data['adjacency_list']}\t{data['distance']}\t{data['backpointer']}")

if __name__ == "__main__":
    reducer()
```
